```python
connection = sqlite3.connect("/Users/josh/Documents/github/josh/spotify_data.db")
cursor = connection.cursor()
conn = sqlite3.connect('your_database_name.db')


query = 'SELECT *  FROM song_metrics WHERE genre != "set()"  '

results = pd.read_sql_query(query, connection)
results['genre'] = results['genre'].str.split(',').str[0]
grouped = results.groupby(['genre', 'year'])
mean_values2 = grouped[['popularity','duration_ms','danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo']].mean().reset_index()

# mean_values2['year'] = pd.to_datetime(mean_values2['year'], format='%Y')
mean_values2['year'] = mean_values2['year'].astype('str')

mean_values2.set_index('year', inplace=True)
mean_values2.head()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
from IPython.display import display


def grouped_bar_chart(genre, category):
    filtered_df = mean_values2[mean_values2['genre'] == genre].reset_index()[['year', category]]
    ax = filtered_df.plot(x='year', y=category, kind='bar', rot=45, figsize=(20, 6))
    plt.xlabel('Year')
    plt.ylabel('Values')
    plt.title(f'Metrics by year for {genre} - {category}')
    plt.show()


genre_dropdown = widgets.Dropdown(
    options=mean_values2['genre'].unique(),
    value=mean_values2['genre'].unique()[0],
    description='Select Genre:',
)

category_dropdown = widgets.Dropdown(
    options=mean_values2.columns[2:],  
    value=mean_values2.columns[2],
    description='Select Category:',
)

interact(grouped_bar_chart, genre=genre_dropdown, category=category_dropdown);
```
